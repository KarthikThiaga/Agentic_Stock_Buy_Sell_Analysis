### Goal: Loop through each number and find whether it is prime or not 
#### What is prime: A number which either divisible by 1 or by its own. 
#### Other way to think this: Any non-prime number is divisible by prime number

```python
prime_list = []
for i in range (1,100):
    if i == 1:
        prime_num_flag = False
    else:
        prime_num_flag = True 
    for j in prime_list:
        if i % j == 0:
            if i != j:
                prime_num_flag = False
    if prime_num_flag == True:
        prime_list.append(i)

print(prime_list)
            
    
```

#### Check if a given string is a palindrome using loops.

```python
string_char = 'Karthik'
lower_str = string_char.lower()
length_of_string = len(lower_str) * -1 
i = 0
is_palindrome = True
for char in lower_str:
    if i > length_of_string: 
        i -= 1
    else:
        i = length_of_string

    if not (char == lower_str[i]):
        is_palindrome = False

if is_palindrome:
    print(f"{string_char} is palindrome")
else:
    print(f"{string_char} is not a palindrome")
    
```

#### Rotate a list by k positions using loops.

```python


# What variables change over time?
#    i = iteratable 
#    j = iteratble 
#    op_list = Output list
# What does each variable represent?
    # List_1 = Input list
    # op_list = output list - Need to keep appending values
    # k = represent number of rotations

# How do they change per iteration?
    # 1st itenration starts from k moves until end
    # 2nd iteration starts from starting of list and moves until k
# When does the loop stop?
    # 1st iteration stops at last value of list
    # 2nd iteration starts from 1st pos and it goes until k
```

```python
List_1 = [1,2,3,4,5,6]
k = 5
op_list = []
for i in range(k,len(List_1)):
    op_list.append(List_1[i])

for j in range(k,0,-1):
    op_list.append(List_1[k - j])

print(op_list)
```

##### Given a sentence, find the word with maximum length using loops.


```python
input_sentence = "The best way to get over your fears is to face your fears"
sentence_list = input_sentence.split()
unqiue_words = set(sentence_list)
max_length = 0
max_word = " "

for words in unqiue_words:
    if len(words) > max_length:
        max_word = words
        max_length = len(words)


print(f'longest word in sentence {input_sentence} is {max_word} and its length was {max_length}')
```
