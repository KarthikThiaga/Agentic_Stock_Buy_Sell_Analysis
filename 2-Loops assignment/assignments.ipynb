```python
### Assignment 12: Factorial Calculation

### Write a program that calculates the factorial of a number input by the user using a while loop.
```

```python
number = int(input("Enter the number: "))
factorial = 1
for i in range(number,1,-1):
    factorial = factorial * i

print(f"Factorial of {number} is: {factorial} ")
```

```python
### Assignment 13: Sum of Digits

## Write a program that calculates the sum of the digits of a number input by the user using a while loop.
```

```python
num = int(input("Enter a number to find sum of digits: "))
dis_num = num
len_of_num = len(str(num))
first_num = 0
total_of_digits = 0

while len_of_num > 1:
    print(len_of_num)
    divisor = 10 ** (len_of_num-1)
    first_num = num//divisor
    print(first_num)
    total_of_digits += first_num
    num = num % divisor
    len_of_num = len(str(num))

final_total = total_of_digits + num
print(f"Sum of digits of {dis_num} is {final_total}")
```

```python
# Get the number from the user as a string to easily process digits
number_str = input("Enter a number: ")

# Initialize the sum of digits and the counter for the while loop
sum_digits = 0
count = 0

# Use a while loop to iterate through each character in the string
while count < len(number_str):
    # Try converting the character to an integer and add it to the sum
    # This also handles cases where the user inputs a non-digit character by skipping it
    try:
        digit = int(number_str[count])
        sum_digits += digit
    except ValueError:
        # Optional: you can print a message if a non-digit is found
        # print(f"Skipping non-digit character: {number_str[count]}")
        pass

    # Increment the counter
    count += 1

# Print the result
print(f"The sum of the digits of {number_str} is: {sum_digits}")
```

```python
### Assignment 14: Prime Number Check

### Write a program that checks if a number input by the user is a prime number using a for loop.
```

```python
number = int(input("Enter a number to find Prime or not: "))

if number > 1: 
    for i in range(1,number):
        if number%i == 0 and number != i:
            isprime = False
        elif number%i == 0 and number == 2:
            isprime = True
        else:
            isprime = True

    if isprime:
        print(f"{number} is prime")
    else:
        print(f"{number} is composite")
elif number == 1:
    print(f"{number} is neither prime nor composite")
else:
    print(f"please enter a valid natural number")
    
```

```python
### Assignment 15: Fibonacci Sequence

### Write a program that prints the first n Fibonacci numbers, where n is input by the user.
```

```python
n = int(input("Enter the number of Fibonacci numbers to print: "))
a, b = 0, 1
count = 0
while count < n:
    print(a)
    a, b = b, a + b
    count += 1
```
