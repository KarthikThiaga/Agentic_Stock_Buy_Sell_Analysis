```python
### Assignment 15: Tuple Functions

### Write functions that take a tuple and return the minimum, maximum, and sum of the elements. Print the results for a sample tuple.

def tuple_agg(input_tupple):
    max_val = 0
    min_val = input_tupple[0]
    sum_of_elements = 0
    for i in input_tupple:
        if i > max_val:
            max_val = i
        elif i < min_val:
            min_val = i
        sum_of_elements = sum_of_elements + i 

    return(max_val,min_val,sum_of_elements)


number_tuples = (4,1,3,5,6,2)

maximum,minimum,sums = tuple_agg(number_tuples)

print(f"Maximum val in tuple", maximum)
print(f"Minimum val in tuple", minimum)
print(f"Sum val in tuple", sums)
```

```python
def min_in_tuple(tpl):
    return min(tpl)

def max_in_tuple(tpl):
    return max(tpl)

def sum_of_tuple(tpl):
    return sum(tpl)

sample_tpl = (1, 2, 3, 4, 5)
print(f"Minimum: {min_in_tuple(sample_tpl)}")
print(f"Maximum: {max_in_tuple(sample_tpl)}")
print(f"Sum: {sum_of_tuple(sample_tpl)}")
```
