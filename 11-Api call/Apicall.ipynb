```python
import requests

public_api = 'https://official-joke-api.appspot.com/random_joke'

try:
    response = requests.get(public_api)
    stat_code = response.status_code
    print(f"Status code output: {stat_code}")
    try:
        joke_dict = response.json()
        print(f"response is {joke_dict} ")
        key_to_search = input("Enter the value from Api you need to see: ")
        try:
            result_of_search = joke_dict[key_to_search]
            print(result_of_search)
        except KeyError:
            print(f'The key entered {key_to_search} is not present try another key')
        except Exception as ex2:
            print(ex2)
    except requests.JSONDecodeError as ex1:
        print(ex1)
except requests.ConnectionError:
    print(f"can't able to connect to {public_api}")
except requests.ConnectTimeout:
    print(f"connection Time out while accessing {public_api}")
except requests.RequestException as ex:
    print(f"exception {ex}")

```

```python
import sys
import os
from dotenv import load_dotenv

try:
    from google import genai
except ModuleNotFoundError:
    print("genai module not installed. Check requirements.")
    sys.exit()

load_dotenv("keys.env")

api_key = os.getenv("API_SECRET_KEY")

if api_key is None:
    print("API key missing in environment.")
    sys.exit()

client = genai.Client(api_key=api_key)

try:
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents={"text": "Why is the sky blue?"}
    )

    if not response or not getattr(response, "text", None):
        print("Empty or malformed response received.")
        sys.exit()

    print(response.text)

except Exception as ex:
    print(f"API call failed: {type(ex).__name__} - {ex}")
```

```python
import sys
import os
from dotenv import load_dotenv
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    handlers=[
        logging.StreamHandler(sys.stdout),
        logging.FileHandler('app.log')
    ]
)
try:
    from google import genai
except ModuleNotFoundError:
    logging.error('genai module not installed. Check requirements.')
    sys.exit()

load_dotenv("keys.env")
logging.info("loaded keys")

api_key = os.getenv("API_SECRET_KEY")

if api_key is None:
    logging.error("API key missing in environment.")
    sys.exit()

client = genai.Client(api_key=api_key)

try:
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents={"text": "Why is the sky blue?"}
    )

    if not response or not getattr(response, "text", None):
        logging.error("Empty or malformed response received.")
        logging.debug(response)
        sys.exit()

    print(response.text)

except Exception as ex:
    logging.error(f"API call failed: {type(ex).__name__} - {ex}")
```
